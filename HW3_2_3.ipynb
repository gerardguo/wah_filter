```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy import signal
from IPython.display import Audio
import ipywidgets as ipw
```

```python
def wah(input, Rp, center_freq, input_gain, fs):
    
    #calculating theta_0 using center frequency provided in Hertz
    theta_0 = center_freq * 2 * np.pi * 1/fs
    
    #setting up b and a
    b = [1, 0, -1]
    a = [1, -2 * Rp * np.cos(theta_0), Rp*Rp]
    
    #apply the filter
    output = signal.lfilter(b, a, input_gain * input)
    
    return output
```

```python
#try it out
guitar, fs = sf.read("Guitar.wav")
guitar_wah = wah(guitar, Rp = 0.99, center_freq = 700, input_gain = 1, fs = fs)
Audio(guitar_wah, rate=fs)
```
